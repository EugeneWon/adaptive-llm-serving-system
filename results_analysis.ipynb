#### baseline_results.csv

- Latency - 정상 
    - batch=1 기준 input_len 32 -> 256 변해도 latency 거의 안 변함 (0.430 -> 0.441s)
    - GPT-2에서 맞는 현상. input은 한 번에 parallel하게 처리되고, bottleneck은 50토큰 autoregressive 생성이라 input 길이 영향이 작음 
- Throughput - 정상, 깔끔하게 스케일링 
    - batch 1 -> 116 tok/s 
    - batch 2 -> 223 tok/s (1.9x) 
    - batch 3 -> 446 tok/s (3.8x)
    - batch 4 -> 860 tok/s (7.4x) <- input_len=32 기준 
    - 배치 늘릴수록 거의 선형으로 증가. 교과서적인 결과 
- GPU util 32~51% - 정상 
    - GPT-2가 너무 작아서 TITAN Xp(12GB)를 못 채우는 거. 의도적인 현상 
- 메모리 - 정상 
    - 모델 자체 ~835MB, batch=8 + input_len=256에서 1657MB. 12BG 대비 여유 있음 
- 주목할 점: batch=4, input_len=256부터 throughput이 조금 꺾이기 시작함 (446 -> 406). input 길이가 길어질수록 KV cache 크기가 커지면서 메모리 대역폭 부담이 생기는 것 - 이게 나중에 최적화할 포인트가 될 수 있음.
